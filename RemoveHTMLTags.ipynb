```python
newfile = open("samples_without_NER_tags.txt","w")
```

```python
import re
# as per recommendation from @freylis, compile once only
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  return cleantext
```

```python
import sys
with open("/content/news_sample_ner.txt") as f:
    for line in f:
        new = cleanhtml(line)
        if not new.isspace():
            newfile.write(new)
            print(new)
```
